```python
import pyspark
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName('spark').getOrCreate()
```

```python
from google.colab import drive
drive.mount('/content/drive/')
```

```python
students=spark.sparkContext.textFile('/content/drive/MyDrive/Students.txt')
```

```python

students.collect()
```

```python
students.map(lambda d:(d.split(' ')[1],1)).reduceByKey(lambda a,b:a+b).collect()
```

```python
students.map(lambda d:(d.split(' ')[2],1)).reduceByKey(lambda a,b:a+b).collect()
```
